```python
from transformers import pipeline
```

```python
summariser=pipeline("summarization")
```

```python
article="""I'm Rudra Pratap Dash, a Third-year student at Maharaja Surjmal Institute of Technology, Janakpuri. I am a budding software engineer with a keen interest in coding and programming and also I am a passionate learner in new technologies.

I have a good understanding of Data Structure and algorithms, Web development, and App Development. Along with all these, my interest also lies in photography, sports, public speaking, and event management.

As a proactive and self-motivated individual, I am always looking for opportunities to collaborate with like-minded individuals and groups and contribute to meaningful projects that have a positive impact on my skills and society.I have keen interest in tech my sepcialization lies in C++ , DSA , HTML , CSS , JS , MongoDB, """
```

```python
summariser(article,max_length=150,min_length=50,do_sample=False)
```
